```python
import matplotlib.pyplot as plt
%matplotlib inline

ranking_means = star_wars.iloc[:, 9:15].mean()

plt.figure(figsize=(10, 5))
plt.bar(star_wars.columns[9:15], ranking_means, color='skyblue')
plt.xticks(rotation=45, ha="right")

plt.ylabel("Averange Rankings (Lower is Better)")
plt.title("Average Rankings of Star Wars Movies")
```

```python
print(ranking_means)
```

Bar chart showes that movies which released first (Episode 4, 5, 6) have better rankings. 

```python
seen_means = star_wars.iloc[:, 3:9].sum()

plt.figure(figsize=(10, 5))
plt.bar(star_wars.columns[3:9], seen_means, color='skyblue')
plt.xticks(rotation=0, ha="center")

plt.ylabel("Number of Watches")
plt.title("Most Viewed Star Wars Movies")
```

Bar chart showes that more people watched movies which released first (Episode 4, 5, 6)
